```python
races = pd.read_excel("illinois_racial.xlsx",header=None).T
races
```

```python
races.columns = races.iloc[0]
races = races.drop(0)
```

```python
races = races.rename(columns={"Total:":"Total Pop", "White alone":"White Pop","Label":"County"})
races.index = pd.Index(range(len(races)))
races
```

```python
races["County"] = [races["County"][i][:-10] for i in range(len(races))]
```

```python
races["County"][1]
```

```python
races["Total Pop"] = [int(races["Total Pop"][i].replace(",","")) for i in range(len(races))]
races["White Pop"] = [int(races["White Pop"][i].replace(",","")) for i in range(len(races))]
races["Black or African American alone"] = [int(races["Black or African American alone"][i].replace(",","")) for i in range(len(races))]
```

```python
races["Percent Not White"] = pd.Series((races["Total Pop"]-races["White Pop"])/races["Total Pop"]*100//1, name="Percent Not White")
races["Percent Black"] = pd.Series((races["Black or African American alone"])/races["Total Pop"]*100//1, name="Percent Not White")
```
